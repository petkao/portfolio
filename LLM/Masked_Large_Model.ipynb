```python
from transformers import BertTokenizer, BertForMaskedLM # Use BertForMaskedLM instead of BertForMaskedML
import torch
```

```python
# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
```

```python
# Input sentenced with a [MASK] token
# text = "The capital of France is [MASK]."
text = "Albert Einstein was born in [MASK]."
inputs = tokenizer(text, return_tensors="pt")
```

```python
# Get model predictions
with torch.no_grad():
  outputs = model(**inputs)
  predictions = outputs.logits
```

```python
# Find the index of [MASK]
mask_token_index = (inputs.input_ids == tokenizer.mask_token_id)[0].nonzero(as_tuple=True)[0]
```

```python
# Get predicted token ID
predicted_token_id = predictions[0, mask_token_index].argmax(axis=-1)
```

```python
# Decode predicted word
predicted_word = tokenizer.decode(predicted_token_id)
```

```python
print(f"Predicted word: {predicted_word}")
```
